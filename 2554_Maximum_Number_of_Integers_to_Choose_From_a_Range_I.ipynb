### Solution 1


```python
class Solution:
    def maxCount(self, banned: List[int], n: int, maxSum: int) -> int:
        banned_set = set(banned)  
        lst=[]
        total=0
        for i in range(1,n+1):
            if i not in banned_set and maxSum>=total+i:
                lst.append(i)
                total+=i
        return len(lst)
```
